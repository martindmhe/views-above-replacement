```python
import pandas as pd
import numpy as np
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, mean_absolute_percentage_error
```

```python
import math
```

```python
# df = pd.read_csv("../data/games.csv")
df = pd.read_csv("./tor_ml_games.csv")
df["is_loss"] = df["is_loss"].astype(int)
df["is_home"] = df["is_home"].astype(int)
df["is_rival"] = df["isRival"].astype(int)
df.head(5)
```

```python
df["total_goals"] = df["leafs_goals"] + df["opponent_goals"]
df["total_penalties"] = df["leafs_penalties"] + df["opponent_penalties"]
df["multi_goal_blown_lead_loss"] = ((df["max_leafs_blown_leads"] > 1) & (df["is_loss"] == 1)).astype(int)

```

```python
training = [
    "is_loss",        
    "goal_differential", 
    "total_goals",     
    "max_leafs_blown_leads", 
    "goals_against_while_leading",
    "leafs_streak",
    # "is_rival"
]
rand = 110

X = df[training]
y = df["views"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.1, random_state=rand
)

model = RandomForestRegressor(
    n_estimators=200,
    random_state=rand,
    n_jobs=-1
)

model.fit(X_train, y_train)

pred = model.predict(X_test)
```

```python
pred = model.predict(X_test)

mape = mean_absolute_percentage_error(y_test, pred)
print("MAPE:", mape)

mae = mean_absolute_error(y_test, pred)
print("MAE:", mae)

# leafs 3 sabres 9 (dec 21, 2023)
# https://www.youtube.com/watch?v=4OGGPx5jgg0 
print(model.predict([[1, -6, 12, 0, 0, -2]]))
```
