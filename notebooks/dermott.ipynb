```python
import pandas as pd
import numpy as np
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
import math
```

```python
# df = pd.read_csv("../data/games.csv")
df = pd.read_csv("./tor_ml_games.csv")
df["is_loss"] = df["is_loss"].astype(int)
df["is_home"] = df["is_home"].astype(int)
df["is_rival"] = df["is_rival"].astype(int)
df.head(5)
```

```python
df["total_goals"] = df["leafs_goals"] + df["opponent_goals"]
df["total_penalties"] = df["leafs_penalties"] + df["opponent_penalties"]
df["multi_goal_blown_lead_loss"] = ((df["max_leafs_blown_leads"] > 1) & (df["is_loss"] == 1)).astype(int)

```

```python
training = [
    "is_loss",        
    "goal_differential", 
    "total_goals",     
    "max_leafs_blown_leads", 
    "goals_against_while_leading",
    "leafs_streak",
    # "is_rival",         
    "fights", 
    # total penalty minutes might be better?
    # "total_penalties", 
    "major_penalties",
    "misconducts",
    "is_home", 
]

X = df[training]
y = df["views"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.1, random_state=23
)

model = RandomForestRegressor(
    n_estimators=200,
    random_state=23,
    n_jobs=-1
)

model.fit(X_train, y_train)
```

```python
pred = model.predict(X_test)

mae = mean_absolute_error(y_test, pred)
print("MAE:", mae)
mse = mean_squared_error(y_test, pred)
rmse = math.sqrt(mse)
print("RMSE:", rmse)
```
